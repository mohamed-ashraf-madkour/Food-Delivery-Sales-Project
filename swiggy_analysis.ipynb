<h2 style="text-align:left; color:#2E86C1;">
Monthly Trend
</h2>

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
sales['Order Month'] = sales['Order Date'].dt.to_period('M').astype(str)
monthly_reveneue = sales.groupby('Order Month')['Price (INR)'].sum()
plt.figure(figsize=(10,5), facecolor="#f7f7f7")
plt.plot(
    monthly_reveneue.index,
    monthly_reveneue.values,
    marker='o',
    linewidth=2,
    color='#2E86C1'
)
plt.title("Monthly Revenue Trend", fontsize=16, fontweight='bold')
plt.xlabel("Month", labelpad=20)
plt.ylabel("Revenue")
plt.grid(True, linestyle='--')
plt.xticks(rotation=45)
for i, v in enumerate(monthly_reveneue.values):
    plt.text(i, v, f"{v:,.0f}", ha='left', va='top', fontsize=9)
plt.tight_layout()
plt.show()
```

<h2 style="text-align:left; color:#2E86C1;">
Daily Trend
</h2>

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
sales['Order day'] = sales['Order Date'].dt.day_name()
daily_reveneue = sales.groupby('Order day')['Price (INR)'].sum()
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
daily_reveneue = daily_reveneue.reindex(order)
plt.figure(figsize=(12,5), facecolor="#f7f7f7")
bars = plt.bar(
    daily_reveneue.index,
    daily_reveneue.values,
    color="#2E86C1"
)
plt.title("Revenue by Day of Week", fontsize=16, fontweight='bold')
plt.xlabel("Day", labelpad=10)
plt.ylabel("Revenue (INR)")
plt.grid(axis='y', linestyle='--', alpha=0.4)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height:,.0f}",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

<h2 style="text-align:left; color:#2E86C1;">
total sales by food type
</h2>

```python
non_veg_keywords=[
    "chicken","egg","fish","mutton",
    "prawn","biryani","kabab","kebab","non-veg","non veg"
]
sales["food category"]=np.where(
    sales["Dish Name"].str.lower().str.contains("|".join(non_veg_keywords),na=False),
    "Non-Veg",
    "Veg"
)
```

```python
total_sales_by_food_type = sales.groupby("food category")["Price (INR)"].sum()

plt.figure(figsize=(6,6))

colors = ['#2E86C1', '#E74C3C', '#F1C40F', '#2ECC71', '#9B59B6']

plt.pie(
    total_sales_by_food_type.values,
    labels=total_sales_by_food_type.index,
    autopct='%1.1f%%',
    startangle=90,
    explode=[0.05]*len(total_sales_by_food_type),
    shadow=True,
    colors=colors[:len(total_sales_by_food_type)]
)

plt.title("Sales Distribution by Food Category", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

<h2 style="text-align:left; color:#2E86C1;">
sales summary performance
</h2> 

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])

sales["Quarters"] = sales["Order Date"].dt.to_period("Q").astype(str)

Quarter_summary = sales.groupby("Quarters", as_index=False).agg(
    Total_Sales = ('Price (INR)', 'sum'),
    Avg_Rating = ('Rating', 'mean'),
    Total_Orders = ('Order Date', 'count')
)
Quarter_summary["Total_Sales"]=Quarter_summary["Total_Sales"].round(2)
Quarter_summary["Avg_Rating"] = Quarter_summary["Avg_Rating"].round(2)

Quarter_summary = Quarter_summary.sort_values("Quarters")

Quarter_summary
```

<h2 style="text-align:left; color:#2E86C1;">
Top Cities By Sales
</h2>

```python
top_cities_by_sales=sales.groupby("City")["Price (INR)"].sum().sort_values().nlargest(5)
plt.figure(figsize=(8,5))

bars = plt.bar(
    top_cities_by_sales.index,
    top_cities_by_sales.values,
    color="#2E86C1"
)

plt.title("Top 5 Cities by Sales", fontsize=14, fontweight='bold')
plt.xlabel("City", labelpad=10)
plt.ylabel("Sales (INR)")
plt.grid(axis='y', linestyle='--', alpha=0.3)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height:,.0f}",
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

<h2 style="text-align:left; color:#2E86C1;">
Total Sales by state 
</h2>

```python
top_Sales_by_state=sales.groupby("State")["Price (INR)"].sum().nlargest(10).sort_values().round(2)
plt.figure(figsize=(8,5))

plt.barh(
    top_Sales_by_state.sort_values().index,
    top_Sales_by_state.sort_values().values,
    color="#2E86C1"
)

plt.title("Top 10 States by Sales", fontsize=14, fontweight='bold')
plt.xlabel("Sales (INR)", labelpad=20)
plt.ylabel("State")

plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

<h2 style="text-align:left; color:#2E86C1;">
Weekly Trend
</h2>

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
sales['Order Weekly'] = sales['Order Date'].dt.to_period('W').astype(str)
weekly_reveneue = sales.groupby('Order Weekly')['Price (INR)'].sum().sort_index()
plt.figure(figsize=(10,5), facecolor="#f7f7f7")
bars = plt.bar(
    weekly_reveneue.index,
    weekly_reveneue.values,
    color="#2E86C1"
)
plt.title("Weekly Revenue Trend", fontsize=16, fontweight='bold')
plt.xlabel("Week",labelpad=10)
plt.ylabel("Revenue (INR)", labelpad=10)
plt.xticks(rotation=75)
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```
